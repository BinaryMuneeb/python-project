##data Cleaning in pandas


```python
import pandas as pd
```

```python
df=pd.read_excel(r"C:\Users\a2z\Desktop\data anyalyst from job\Customer Call List.xlsx")
```

```python
df
```

```python
df1=df.drop_duplicates()
```

```python
df1
```

```python
df1=df1.drop(columns="Not_Useful_Column")
```

```python
df1
```

```python
# df1["Last_Name"].str.lstrip('...')

# df1["Last_Name"].str.lstrip('/')
# 
df["Last_Name"]=df1["Last_Name"].str.rstrip('_')

# df1["Last_Name"].str.lstrip('...')
```

```python
df1
```

```python
df["Last_Name"]= df1["Last_Name"].str.rstrip('_')
```

```python
df["Last_Name"]= df1["Last_Name"].str.lstrip('/')
```

```python
df["Last_Name"].str.strip('123._/')
```

```python
df
```

```python
df=df.drop(columns='Not_Useful_Column')
```

```python
df
```

```python
df.drop_duplicates()
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]', '', regex=True)
```

```python
df
```

```python
df
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
```

```python
df
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
df
```

```python

df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')

df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')
df
```

```python
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',', expand=True)
df
```

```python
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace('Yes','Y')
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace('No','N')
```

```python
df
```

```python
df["Paying Customer"]=df["Paying Customer"].str.replace('Yes','Y')
df["Paying Customer"]=df["Paying Customer"].str.replace('No','N')
```

```python
df
```

```python
df=df.replace('N/a',' ')
```

```python
df
```

```python
df = df.replace('NaN',' ')
```

```python
df
```

```python
df=df.fillna('')
df
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)

df
```

```python
for x in df.index:
    if df.loc[x, "Paying Customer"] == ' ':
        df.drop(x, inplace=True)

df
```

```python
df.drop_duplicates()
```

```python
new_df=df.reset_index(drop=True)
```

```python
new_df
```

```python
df
```

```python
new_df = new_df[new_df['Phone_Number'].str.strip() != '']
```

```python
new_df
```

```python
csv_file_path = 'cleaned_data.csv'
new_df.to_csv(csv_file_path, index=False)

print(f"DataFrame has been successfully saved to {csv_file_path}")
```
